Импорт библиотек и считывание данных из файла:

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

data = pd.read_csv('STEPS.csv', sep=',')
X = data['Date']
Y = data['Steps']
Y
```

Расчет основных параметров набора данных (среднее, медиана, СКО, дисперсия):

```python
mean = Y.mean()
sko = Y.std()
dispersion = sko**2
median = Y.median()
print(f"Математическое ожидание: {mean}")
print(f"Медиана: {median}")
print(f"Стандартное отклонение (СКО): {sko}")
print(f"Дисперсия: {dispersion}")
```

С помощью matplotlib.pyplot отобразим полученные значения на графике

```python
plt.plot(X, Y, linewidth = 2)
plt.axhline(y = mean, color = "r", linewidth = 1)
plt.axhline(y = median, color = "y", linewidth = 1)
plt.axhline(y = mean - 3*sko, color = "g", linewidth = 1, linestyle = "--")
plt.axhline(y = mean + 3*sko, color = "g", linewidth = 1, linestyle = "--")
plt.xlabel("Date")
plt.ylabel("Steps")
plt.legend(["Шаги", "Математическое ожидание", "Медиана", "Интервалы 3σ"])
plt.show

```

Построим гистограмму и кривую Гаусса:

```python
plt.hist(Y, density = True)
y = np.linspace(Y.min(), Y.max())
gaus = norm.pdf(y, mean, sko)
plt.plot(y, gaus, label='Кривая Гаусса')
plt.xlabel('Шаги')
plt.ylabel('Плотность')
plt.legend()
plt.show()
```
